# Comparison Operators

```python
a = 10
b = 20
```

```python
print(a == b)   # False
print(a != b)   # True
print(a > b)    # False
print(a < b)    # True
print(a >= 10)  # True
print(b <= 10)  # False
```

# ogical Operators

```python
x = True
y = False

print(x and y)   # False
print(x or y)    # True
print(not x)     # False
```

# Bitwise Operators

```python
a = 5     # 0101
b = 3     # 0011

print(a & b)   # 1   (0001)
print(a | b)   # 7   (0111)
print(a ^ b)   # 6   (0110)
print(~a)      # -6
print(a << 1)  # 10  (shift left)
print(a >> 1)  # 2   (shift right)
```

# Membership Operators

```python
text = "python"

print("p" in text)       # True
print("z" in text)       # False
print("py" not in text)  # False
```

# Identity Operators

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)      # True  (same memory)
print(a is c)      # False (different objects)
print(a is not c)  # True
```
